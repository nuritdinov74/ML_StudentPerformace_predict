```python
df = pd.get_dummies(df, columns=['race_ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course'])
```

```python
df['gender'] = (df['gender'] == 'male').astype(int)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

features = df.drop(columns=['math_score','all_score', 'average_score', 'Unnamed: 0'])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)
```

```python
df.columns
```

```python
scaled_data
```

```python
df.shape
```

```python
sns.boxplot(df['average_score'])
plt.show()
```

```python
x = scaled_data
y = df['math_score']
```

```python
x.shape
```

```python
from sklearn.model_selection import  train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python

from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

r2 = r2_score(y_test, y_pred)
```

```python
import joblib

joblib.dump(model, 'model_3.pkl')
joblib.dump(scaler, 'scaler_3.pkl')
```

```python
print("R2 score:", r2)
```
